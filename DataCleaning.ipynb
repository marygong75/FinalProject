```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
suicide_df = pd.read_csv("who_suicide_statistics.csv")
suicide_df.head()
```

```python
clean_df = suicide_df.loc[suicide_df['country'] == "United States of America"]
clean_df
```

```python
clean_df.dtypes
```

```python
age_df = clean_df.loc[clean_df['age'] == "25-34 years"]
age_df
```

```python
age_df.drop(age_df.columns[[3]], axis=1, inplace=True)
age_df
```

```python
#sort the data and save it as a new file
age_df.sort_values(by='suicides_no').set_index('suicides_no').reset_index().to_csv('who_suicide_statistics_clean.csv', index=False)
```
